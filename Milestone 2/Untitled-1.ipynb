```python
import pandas as pd
import openpyxl
```

```python
wb = openpyxl.load_workbook(r'./data.xlsx') 
year_list = wb.sheetnames
year_list.reverse()
print(year_list)

df_1 = pd.read_excel('./data.xlsx', sheet_name='2009')
print(df_1)
name_list = df_1['City'].tolist()
```

```python
df_save = pd.DataFrame()
df_save['City'] = name_list

for year in year_list:
    df = pd.read_excel('./data.xlsx', sheet_name=year)
    index_dict = {}
    for i in range(len(df)):
        index_dict[df.loc[i, 'City']] = df.loc[i, 'Cost of Living Index']

    df_save[year] = 0
    for j in range(len(df_save)):
        df_save.loc[j, year] = index_dict.get(df_save.loc[j, 'City'])

print(df_save)

```

```python
df_save.to_csv('test.csv', index=False)
```
